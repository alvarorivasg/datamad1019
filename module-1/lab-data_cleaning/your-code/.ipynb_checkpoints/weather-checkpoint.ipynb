## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
data=pd.read_csv('/Users/alvarorivas/desktop/code/ironhack/datamad1019/module-1/lab-data_cleaning/weather-raw.csv')
data
```

```python
datamax=data[data['element']=='tmax']
datamax
```

```python
datamin=data[data['element']=='tmin']
datamin
```

```python
newdat=datamax.melt(id_vars=['id','year','month','element'],var_name='Día',value_name='tmax')
newnewdat=newdat[['year','month','Día','tmax']]
display(newnewdat.head(20))
```

```python
newdar=datamin.melt(id_vars=['id','year','month','element'],var_name='Día',value_name='tmin')
newnewdar=newdar[['year','month','Día','tmin']]
display(newnewdar.head(20))
```

```python
megatable=pd.merge(newnewdat,newnewdar)
megatable.head(15)
```

```python
filtered = megatable[(megatable['tmax'].notnull()) & (megatable['tmin'].notnull())]
filtered.head()
```

```python
seda=filtered.sort_values(by=['month'])
seda.head()
```

```python
seda['Día']=seda['Día'].str.replace('d','')
seda.head()
```

```python
seda['Día']=seda['Día'].astype('int64')
```

```python
sedon=seda.sort_values(by=['month','Día'])
sedon.head()
```

```python
avgmax=sedon['tmax'].mean()
avgmax
#ya podría sacar lo que me venga en gana
```

```python
minmin=sedon['tmin'].min()
minmin
```

```python
avg_month = sedon.groupby('month', as_index=True).agg({'tmax':'mean','tmin':'mean'})
avg_month
```
