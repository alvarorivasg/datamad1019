```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data.csv")
data.head()
```

```python
data.cut = data.cut.replace({'Ideal':1, 'Premium':2, 'Very Good':3, 'Good':4, 'Fair':5})
data.color = data.color.replace({'D':1, 'E':2, 'F':3, 'G':4, 'H':5, 'I':6, 'J':7})
data.clarity = data.clarity.replace({'IF':1,'VVS1':2,'VVS2':3, 'VS1':4, 'VS2':5, 'SI1':6, 'SI2':7, 'I1':8})
```

```python
data.head()
```

```python
data.columns
```

```python
columns=['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']
dictio={}
for col in columns:
    dictio[col]=[np.mean(data[col]),np.std(data[col])]
    data[col] = (data[col] - np.mean(data[col])) / np.std(data[col]) 
data.head()
```

```python
dictio
```

```python
from h2o.automl import H2OAutoML
```

```python
import h2o
h2o.init(nthreads = -1, max_mem_size = 4)
```

```python
datah2o=h2o.H2OFrame(data)
```

```python
y_columns="price"
x_columns=['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']
```

```python
train, test=datah2o.split_frame(ratios = [.8])
```

```python
aml = H2OAutoML(max_runtime_secs= 180,max_models= 15, seed= 1, nfolds=0)
aml.train(x = x_columns, y = y_columns, training_frame = train, validation_frame=test)
```

```python
lb= aml.leaderboard
lb
```

```python
datatest=pd.read_csv("test.csv")
datatest
```

```python
datatest.cut = datatest.cut.replace({'Ideal':1, 'Premium':2, 'Very Good':3, 'Good':4, 'Fair':5})
datatest.color = datatest.color.replace({'D':1, 'E':2, 'F':3, 'G':4, 'H':5, 'I':6, 'J':7})
datatest.clarity = datatest.clarity.replace({'IF':1,'VVS1':2,'VVS2':3, 'VS1':4, 'VS2':5, 'SI1':6, 'SI2':7, 'I1':8})
```

```python
predfinal = aml.leader.predict(datah2otest)
pdpred = h2o.as_list(predfinal)
```
